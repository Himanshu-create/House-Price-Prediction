```python
#lets remove the outliers from the location and the rate
def remove_rate_outliers1(df):
    df_out = pd.DataFrame()
    for key , subdf in df.groupby("location"):
        m = np.mean(subdf["rate"])
        st = np.std(subdf["rate"])
        reduced_df = subdf[(subdf["rate"]>(m-st)) & (subdf["rate"]<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index = True)
    return df_out

train1 = remove_rate_outliers1(data)
print(train1.shape)
train1.head()
```

```python
plt.scatter(train1["sqft"],train1["price"])
```

```python
def remove_rate_outliers2(df):

    df_out = pd.DataFrame()
    for k, subdf in df.groupby("location"):
        q1 = subdf["rate"].quantile(0.25)
        q3 = subdf["rate"].quantile(0.75)
        iqr = q3 - q1
        reduced_df = subdf[(subdf["rate"]>(q1 - 1.5*iqr)) & (subdf["rate"] < (q3 + 1.5*iqr))]
        df_out = pd.concat([df_out,reduced_df],ignore_index = True)
    return df_out

train2 = remove_rate_outliers2(data)
print(train2.shape)
train2.head()
```

```python
def scatter_plot(df,location):
    bhk2 = df[(df["location"] == location) & (df["size"] == 2)]
    bhk3 = df[(df["location"] == location) & (df["size"] == 3)]
    plt.scatter(bhk2.sqft,bhk2.price,color = "blue",label = "2BHK")
    plt.scatter(bhk3["sqft"],bhk3["price"],color = "green",label = "3BHK")
    
    plt.xlabel("SQFT")
    plt.ylabel("Price")
    plt.title(location)
    plt.legend()

scatter_plot(train,"Hebbal")
```

```python
train_cp = train.copy()
#lets remove the data which 3bhk have less price for the same square feet
def remove_bhk_outlier(df):
    exclude = np.array([])
    for location,location_df in df.groupby("location"):
        bhk_stats = {}
        for bhk , bhk_df in location_df.groupby("size"):
            bhk_stats[bhk] = {
                "mean" : np.mean(bhk_df["rate"]),
                "std" : np.std(bhk_df["rate"]),
                "count" : bhk_df.shape[0]
                }
        
        for bhk,bhk_df in location_df.groupby("size"):
            stats = bhk_stats.get(bhk - 1)
            if stats and stats["count"] > 5:
                exclude = np.append(exclude,bhk_df[bhk_df["rate"]<(stats["mean"])].index.values)
    return df.drop(exclude,axis = "index")

train = remove_bhk_outlier(train)
train.shape
```

```python
train.drop(["rate","area"],axis = 1,inplace = True)
#pandas dummy method
locations = pd.get_dummies(train["location"])
locations.head(3)
```

```python
training = pd.concat([train.drop(["location","area_type"],axis = 1),locations.drop(['Yeshwanthpur'],axis = 1)],axis = "columns")
```

```python
training.head(3)
```

```python
X = training.drop(["price"],axis = 1)
Y = training["price"]
```

```python
#standard scale:
from sklearn.preprocessing import StandardScaler as ss

ss_sqft = ss()
X[["sqft"]] = ss_sqft.fit_transform(X[["sqft"]])

ss_price = ss()
Y = np.array(Y)
Y = Y.reshape((Y.shape[0],1))
Y = ss_price.fit_transform(Y)
Y = pd.DataFrame(Y)
```

```python
from sklearn.model_selection import train_test_split as tts
X_train,X_test,Y_train,Y_test = tts(X,Y,test_size = 0.2,random_state = 10)
```

```python
###linear regressor
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
linear.fit(X_train,Y_train)
linear.score(X_test,Y_test)
```

```python
#####SVE
from sklearn.svm import SVR
regressor = SVR()
regressor.fit(X_train,Y_train)
regressor.score(X_test,Y_test)
```

```python
######Ramdomforest
from sklearn.ensemble import RandomForestRegressor as rfr
regressor = rfr(n_estimators = 10)
regressor.fit(X_train,Y_train)
regressor.score(X_test,Y_test)
```

```python
#cross k fold
from sklearn.model_selection import ShuffleSplit, cross_val_score

cv = ShuffleSplit(n_splits = 5,test_size = 0.2,random_state = 0)
cross_val_score(regressor,X,Y,cv = cv)
```

```python
cross_val_score(LinearRegression(),X,Y,cv = cv)
```
